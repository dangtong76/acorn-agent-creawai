### 1. 데코레이터란 무엇이냐?

#### 1.1 인자없는 함수에 대한 데코레이터

```python
def logger(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper
```

```python
@logger
def say_hello():
    print("Hello!")

@logger
def say_goodbye():
    print("goodbye!")
```

```python
say_goodbye()
```

#### 1.2 인자가 있는 함수에 대한 데코레이터

args : 인자 , 매개변수 logger(1,2,3) 

kargs : 키워드 인자 logger(a=1, b=2, c=3)

```python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} function with {args} {kwargs}")
        result = func(*args, **kwargs)
        print(f"{func.__name__} returned {result}")
        return result
    return wrapper
```

```python
@logger
def add(a,b):
    return a+b 
```

```python
add(3,5)
```

### 3. 자료구조 변환

#### 3.1 dict -> 객체

```python
class User:
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email
```

```python
data = {
    "name": "dangtong",
    "age": 49,
    "email": "dangtong@gmail.com"
}
```

```python
user = User(**data)
### User(self, name=dangtong, age=49, email=dangtong@gmail.com)
```

```python
print(user.name, user.age, user.email)
```

#### 3.2 객체 -> DICT

```python
dictionary = user.__dict__
print(dictionary)
```

#### 3.3 dict, json <-> pydantic Model (실무에서 가장 많이 쓴다)

```python
from pydantic import BaseModel
```

```python
class User(BaseModel):
    name: str
    age: int
    email: str
```

```python
data = {"name": "park", "age": 33, "email": "park@gmail.com"}
```

```python
# dict -> pydantic model
user = User(**data)
```

```python
# model -> dict
print(user.model_dump())
```

```python
# model -> json
print(user.model_dump_json())
```
